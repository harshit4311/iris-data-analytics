## Advanced Feature Engineering & Predictive Modeling

```python
""" setup """

# Import required libraries
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score, KFold
import numpy as np

print('import done')
```

## 1. Feature Engineering

#### 1.1 Dimensionality Reduction with PCA




```python
from sklearn.preprocessing import StandardScaler

# standardise all the features before doing PCA
X = df[iris.feature_names]
X_scaled = StandardScaler().fit_transform(X)

# apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# plot variance
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('No.of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.grid(True)
plt.show()
'''
```

```python
"""

Observation:
- The first 2 components explain >95% of the variance, which is sufficient --> we will reduce to 2D

"""

pca = PCA(n_components=2)
X_pca_final = pca.fit_transform(X_scaled)

# visualise
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca_final[:, 0], y=X_pca_final[:, 1], hue=df['target_name'])
plt.title("PCA - 2 Components Visualization")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```

#### 1.2 Feature Selection

```python
# (a) Filter Method  Correlation + SelectKBest

# using 'SelectKBest' with 'ANOVA F-test'
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, df['target_encoded'])

selected_features = np.array(iris.feature_names)[selector.get_support()]
print("Filter Method - Selected Features:", selected_features)
```

```python
# (b) Wrapper Method  Recursive Feature Elimination (RFE)

model = LinearRegression()
rfe = RFE(model, n_features_to_select=2)
fit = rfe.fit(X, df['target_encoded'])

selected_rfe_features = np.array(iris.feature_names)[fit.support_]
print("Wrapper Method - RFE Selected Features:", selected_rfe_features)
```

```python
"""

Observations:

- filter method selected = [petal length (cm), petal width (cm)]
- wrapper method selected = [petal length (cm), petal width (cm)]

And these will be used for different models to observe performance impact

"""
```

#### 1.3 Handling Imbalanced Data

```python
print("Class distribution:\n", df['target_name'].value_counts())
```

```python
"""
Observations:

- The class distribution is balanced in our Iris dataset (50-50-50)
- So, there is no need of oversampling/under-sampling or handling more imbalanced data

"""
```

## 2. Predictive Modelling

```python
""" splitting the data for training/testing and model, both"""

# using the earlier selected features from filter method
X_selected = df[['petal length (cm)', 'petal width (cm)']]
y = df['target_encoded']

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

#### 2.1 Linear Regression Model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("Linear regression:")
print("MAE:", mean_absolute_error(y_test, y_pred_lr))
print("MSE:", mean_squared_error(y_test, y_pred_lr))
print("R-squared:", r2_score(y_test, y_pred_lr))
```

#### 2.2 Decision Tree Model

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

print("Decision tree accuracy:", accuracy_score(y_test, y_pred_dt))
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))

# visualize the decision tree
plt.figure(figsize=(10,6))
plot_tree(dt, feature_names=X_selected.columns, class_names=iris.target_names, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

## 3. Model Evaluation

#### 3.1 Compare Models

```python
"""
Observations:

- Linear regression is not ideal for classification tasks as its predictions are continuous
- Decision tree provides discrete class outputs & it performed better in classification metrics

"""
```

#### 3.2 Cross-Validation

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Linear regression CV [using regression metrics]
lr_scores = cross_val_score(lr, X_selected, y, scoring='r2', cv=kf)
print("Linear Regression - R2 Scores:", lr_scores)
print("Mean R2:", lr_scores.mean())
print("Std Dev:", lr_scores.std())

# Decision tree CV [using classification accuracy]
dt_scores = cross_val_score(dt, X_selected, y, scoring='accuracy', cv=kf)
print("\nDecision Tree - Accuracy Scores:", dt_scores)
print("Mean Accuracy:", dt_scores.mean())
print("Std Dev:", dt_scores.std())
```
