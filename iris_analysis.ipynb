# 1. Data acquisition

```python
''' 1.1 Download the dataset '''

from sklearn.datasets import load_iris
import pandas as pd

# load the DS from sklearn
iris = load_iris()
```

```python
''' 1.2 Convert the dataset into a DataFrame '''

# create a DF from data & feature names
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['target_name'] = df['target'].map(lambda x: iris.target_names[x])
```

```python
''' 1.3 Display the first and last five records '''

print("First 5 records:")
display(df.head())

print("Last 5 records:")
display(df.tail())
```

```python
''' 1.4 Display column headings, data types, and summary stats '''

print("Column headings:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

print("\nDataset description:")
print(iris.DESCR)

print("\nStatistical summary:")
display(df.describe())
```

```python
''' 1.5 Observations '''


'''
My Observations:

- No.of features(excluding target): 4

- No.of examples: 150

- Feature names: sepal length, sepal width, petal length, petal width

- Types of data attributes:

- Features: Numeric(float)

- Target: Categorical(string via `target_name`)

'''
```

```python
df.describe()
```

# 2. Preparing data

```python
''' 2.1 Check for duplicates, missing values, inconsistencies, and outliers '''

print("Duplicate rows:", df.duplicated().sum())
print("Missing values:")
print(df.isnull().sum())

# boxplots for detecting outliers
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 8))
sns.boxplot(data=df.drop(['target', 'target_name'], axis=1))
plt.title("boxplot for detecting outliers")
plt.show()
```

```python
''' 2.2 Cleaning the data '''

# remove duplicates if there are any
df = df.drop_duplicates()

''' 
My observations/analysis:
Turns out, there are no missing values or inconsistencies in our dataset/dataframe 
Also, there are no clear outliers which need to be removed + the dataset is small
'''
```

```python
''' 2.3 Encode data into categories: 1 or 0 '''

# encode 'target_name' using label encoding (done in the 'target' col.)
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()

df['target_encoded'] = le.fit_transform(df['target_name'])
```

```python
''' 2.4 Observations '''

'''
- Duplicates: removed if any (none in our case)
- Missing Values: not found any
- Inconsistencies: not found any; data is very clean
- Outliers: not requried
- Encoding Method: used "Label Encoding" for `target_name` col.
'''
```

# 3. Data exploration + visualisation

```python
''' 3.1 Scatter plots for each feature vs target '''

features =iris.feature_names
plt.figure(figsize=(12, 12))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.scatterplot(data=df, x=feature, y='target_encoded', hue='target_name')
    plt.title(f"{feature} vs TARGET")
plt.tight_layout()
plt.show()
```

```python
''' 3.2 Additional Visualizations '''

# (a) Pair plot
sns.pairplot(df[features + ['target_name']], hue='target_name')
plt.suptitle("Pair plot of Iris features", y=1)
plt.show()
```

```python
'''
Justification:

Pair plots help us in visualising relationships & class separation across ALL the feature combinations
This somewhat aids in identifying which features are the most effective in distinguishing between classes (like-- petal length & petal width are good parameters/separators)

'''
```

```python
# (b) Correlation Heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title("Feature correlation heatmap")
plt.show()
```

```python
'''
Justification:

A heatmap allows us to analyze the strength of linear relationships between features
For example, petal length & petal width show HIGH correlation which indicate potential redundancy OR combined predictive power

'''
```
